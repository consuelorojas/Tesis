```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import waves_simulation as ws


sns.set_style("darkgrid")
```

```python
signal = ws.SimulatedSignal(10)

plt.plot(signal.faradayWave(), label = 'Faraday Wave')
plt.xlabel('Time (s)')
plt.ylabel('Surface Height (m)')
plt.title('1D Faraday Wave Simulation')
plt.xlim(0,1000)
plt.show()
```

```python
noise = signal.noise()

plt.plot(noise, label = 'noise')
plt.xlabel('Time (s)')
plt.ylabel('Surface Height (m)')
plt.title('Noise Simulation')
plt.xlim(0,1000)    
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2,1, sharey=True)
ax1.plot(signal.signal(), label = 'signal')
ax1.plot(signal.faradayWave(), label = 'Faraday Wave', alpha = 0.3)
ax1.plot(signal.noise(), label = 'noise', alpha = 0.5)

#ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Surface Height (m)')
ax1.set_title('1D Faraday Wave Simulation')
ax1.set_xlim(0,1000)

ax2.plot(signal.signal(), label = 'signal')
ax2.plot(signal.faradayWave(), label = 'Faraday Wave', alpha = 0.3)
ax2.plot(signal.noise(), label = 'noise', alpha = 0.5)

ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Surface Height (m)')
#ax2.set_title('1D Faraday Wave Simulation')
ax2.set_xlim(0,200)

plt.show()
```
